```python
#dependencies & setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sunroof_city = "project-sunroof-city-09082017.csv"
```

```python
city_data = pd.read_csv(sunroof_city)
city_data.sort_values('existing_installs_count', ascending = False).head(10)
```

```python
new_city_data = city_data.iloc[:,[0,1,9,18,20,29,30]]
new_city_data.dropna()
```

```python
new_city_data.dropna(how='all').sort_values('count_qualified', ascending = False)
new_city_data.dropna()
```

```python
new_city_data.sort_values('existing_installs_count', ascending = False)
```

```python
St_Louis = city_data.loc[new_city_data["region_name"] == "St. Louis"]
St_Louis.info()
```

```python
directions = ["sunlight_north", "sunlight_south", "sunlight_east", "sunlight_west","sunlight_flat"]
members = [6.763237e+07, 2.918317e+08, 2.537756e+08, 2.083540e+08, 1.285719e+09]

colors = ["yellowgreen", "red", "lightcoral", "lightskyblue","orange"]
explode = (0.05, 0, 0, 0,0)
```

```python
plt.title("directions of sunlight")
plt.pie(members, explode=explode, labels=directions, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
plt.savefig("Images/direction of sunlight.png")
plt.show()
```

```python
group_state = new_city_data.groupby(["state_name"])
print(group_state.head())
```

```python
St_Louis = group_state.loc["Missouri"]
St_Louis
```

```python
narrowed_light_data2 = state_data[["state_name", "existing_installs_count", "count_qualified", "yearly_sunlight_kwh_total"]]
narrowed_light_data2
```

```python
narrowed_light_data2_index = narrowed_light_data2.set_index("state_name")
narrowed_light_data2_index_sort = narrowed_light_data2_index.sort_values('existing_installs_count', ascending = False)
top_25 = narrowed_light_data2_index_sort.head(10)
top_25
```

```python
plt.figure(figsize=(10,7))
test1=np.array(top_25['count_qualified'])
test2=np.array(top_25['yearly_sunlight_kwh_total']/10000000000000)
test3=np.array(top_25['existing_installs_count'])
variablelabel = np.array(narrowed_light_data2.state_name)  

plt.scatter(x=test3,y=test1,s=test2,
           color= 'lightcoral', edgecolor='k',
           label= 'sunlight', alpha =.6, linewidth= .75)

for i, txt in enumerate(variablelabel):
   plt.annotate(txt, (test3[i], test1[i]))


plt.title('Relationship between Sunlight total and Counts Qualified and Counts existing', fontsize=20)
plt.xlabel('Number of Existing Solar Pannel Installations')
plt.ylabel('Number of buildings qualified for solar')
plt.xlim(0, max(test3))
plt.ylim(0,max(test1))

# Create a legend

legend = plt.legend(loc = "best", fontsize= 10, numpoints =1,
                   scatterpoints=1, title= 'Sunlight Relationship', labelspacing = .5)
# make bubbles the same size
#legend.legendHandles[0]._sizes = [50]
#legend.legendHandles[1]._sizes = [50]
#legend.legendHandles[2]._sizes = [50]

#reference regarding cicle size
plt.text(2001,35000, "Note: Circle size correlates with amount of light per state (*10e7).")
plt.grid(b=None, which='major', axis='both')
#plt.savefig('SunlightStateBubble.png')
plt.show()
# Save Figure
```
